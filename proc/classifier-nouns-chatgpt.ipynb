```python
import openai

client = openai.OpenAI(api_key=""
)
import pandas as pd

file_path = "/Users/reneesoler/Documents/Data Literacy Project/Nouns to classify.xlsx"
df = pd.read_excel(file_path)

```

```python
if "lemma" not in df.columns:
    raise ValueError("The expected column 'lemma' is not found in the Excel file.")

nouns = df["lemma"].tolist()

def classify_nouns_batch(nouns_batch):
    batch_prompt = (
        "Classify the following German nouns into one of four categories: "
        "'plant', 'animal', 'beruf' (profession), or 'none'.\n"
        "Format the response as a list where each line is '<noun>: <category>'\n\n"
        + "\n".join(nouns_batch)
    )

    response = client.chat.completions.create(
        model="gpt-3.5-turbo",
        messages=[{"role": "system", "content": batch_prompt}],
        max_tokens=1500
    )

    
    classifications = {}
    lines = response.choices[0].message.content.strip().split("\n")
    for line in lines:
        parts = line.split(": ")
        if len(parts) == 2:
            noun, category = parts
            classifications[noun.strip()] = category.strip().lower()

    return classifications
```

```python



if "Category" not in df.columns:
    df["Category"] = None

batch_size = 50

for i in range(0, len(nouns), batch_size):
    batch = nouns[i:i + batch_size]
    
    
    batch_to_classify = [word for word in batch if pd.isna(df.loc[df["lemma"] == word, "Category"]).any()]
    
    if batch_to_classify:  
        batch_results = classify_nouns_batch(batch_to_classify)

    
        for noun, category in batch_results.items():
            df.loc[df["lemma"] == noun, "Category"] = category

       
        if (i // batch_size) % 10 == 0:
            df.to_excel("/Users/reneesoler/Documents/Data Literacy Project/Classification chat gpt.xlsx", index=False)
            print(f"Processed {i + len(batch_to_classify)} / {len(nouns)} nouns...")

```
